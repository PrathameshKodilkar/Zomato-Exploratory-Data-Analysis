```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_values=final_df.Country.value_counts().values
country_values
```

### Pie Chart

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

### Observation :
1. Zomato maximum records are from India.
2. After that USA and UK.

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### Observations :
1. Whenever rating is between 4.5 to 4.9 it indicates Excellent.
2. From 3.5 to 3.9 it is Good.
3. If rating between 4.0 to 4.4 it is Very Good.
4. From 3.0 to 3.4 it is Average Rating.
5. When it is not Rated it shows White color and Holds the maximum count.

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(15,6)
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
